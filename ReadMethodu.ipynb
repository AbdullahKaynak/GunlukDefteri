# Metin Dosyaları okuma
### **r**: Dosyaları okumak için okuma modu
### **w**: Dosyaları yazmak için yazma modu

```python
## dosyamızı okuyalım

example1 = "example1.txt"
file1 = open(example1, "r")
```

# Dosya Açmanın Daha İyi Bir Yolu

### With ile dosyayı açarsan dosyayla işin bittiğinde kendini otomatik kapatır. bu yüzden with kullanılır

```python
# dosyayı şu şekilde aç her zaman

with open(example1, "r") as file1:
    fileContent = file1.read()
    print(fileContent)
```

## Dosyanın tamamını okumak zorunda değiliz, örneğin, .read() metoduna parametre olarak üç karakter girerek ilk 4 karakteri okuyabiliriz:

```python
with open(example1, "r") as file1:
    print(file1.read(4))                ## ilk 4 karakteri okur ve kapatır
    print(file1.read(4))                ## ilk 4 karakteri okur ve kapatır
    print(file1.read(7))                ## ilk 7 karakteri okur ve kapatır
    print(file1.read(15))                ## ilk 15 karakteri okur ve kapatır
```

## Dosyanın her satırını tek tek okumak için readline() yöntemini de kullanabiliriz:

```python
with open(example1, "r") as file1:
    print("first line: " + file1.readline())
```

```python
# Ayrıca, okumak istediğimiz karakter sayısını belirtmek için readline() fonksiyonuna bir argüman da geçirebiliriz.
# Ancak, read() fonksiyonunun aksine, readline() en fazla bir satır okuyabilir.

with open(example1, "r") as file1:
    print(file1.readline(20))       ## Satır sonundan sonrasını okumaz
    print(file1.read(20))           ## Sonraki 20 karakteri döndürür
```

## Her satırı tek tek incelemek için bir döngü kullanabiliriz:

```python
with open(example1, "r") as file1:
    i = 0;
    for line in file1:
        print("satır", str(i), ": ", line)
        i = i+1
```

## `readlines()` yöntemini kullanarak metin dosyasını bir listeye kaydedebiliriz:

```python
## # Tüm satırları oku ve liste olarak kaydet

with open(example1, "r") as file1:
    fileasList = file1.readlines()
```

```python
print(fileasList[0])
print(fileasList[1])
print(fileasList[2])
```
