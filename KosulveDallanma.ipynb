#### "Ve" ifadesi yalnızca her iki koşul da doğru olduğunda doğrudur. (and)
#### "Veya" ifadesi, bir koşul veya her ikisi de doğru olduğunda doğrudur. (or)
#### "Değil" ifadesi ise bunun tersi doğruluk değerini verir. (not)

## "and"

```python
album_year = 1980

if(album_year > 1979) and (album_year < 1990):
    print ("Albümün çıkış yılı 1980 ile 1989 yılları arasındaydı.")
    
print("")
```

## "or"

```python
album_year = 1990

if(album_year < 1980) or (album_year > 1989):
    print ("Albüm 1980'lerde yapılmadı.")
else:
    print("Albüm 1980'lerde yapıldı. ")
```

## "not"

```python
album_year = 1983

if not (album_year == 1984):
    print ("Albümün yılı 1984 değil.")
```

### Soru 1 - Lionel Messi adlı oyuncunun 10'dan fazla başarıya sahip olup olmadığını kontrol eden bir Python programı yazın. Koşul doğruysa, oyuncunun adını, spor dalını ve başarılarını yazdırın; aksi takdirde "10'dan fazla başarısı yok" yazdırın.

```python
player_name = "Lionel Messi"
sport = "Soccer"
achievements = 11
```

```python
if player_name== "Lionel Messi":
    if achievements >= 10:
        print(f"{player_name} adlı {sport} oyuncusunun {achievements} başarısı vardır.")
    else:
        print(f"{player_name} adlı {sport} oyuncusunun 10'dan fazla başarısı yoktur.")
else:
    print("Lütfen ismi doğru giriniz.")
```
