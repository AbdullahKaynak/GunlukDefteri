# İçeriği bir dosyadan diğerine kopyalama

```python
# kaynak dosyasını okuma modunu aç
with open(new, "r") as file1:
    ## hedef dosyayı yazma modunu aç
    with open("destination.txt","w") as destination_file:
        ## kaynak dosyadan satırları oku ve hedef dosyaya kopyala
        for line in file1:
            destination_file.write(line)

    # 'with' bloğu sona erdiğinde hedef dosya otomatik olarak kapatılır
# 'with' bloğu sona erdiğinde kaynak dosya otomatik olarak kapatılır
```

### x+ : Özel oluşturma ve okuma/yazma modu. Okuma ve yazma için yeni bir dosya oluşturur, ancak zaten mevcutsa bir hata oluşturur.
### a+ : Ekle ve okuma modu. Hem ekleme hem de okuma için bir dosya açar. Mevcut değilse dosyayı oluşturur.
### w+ : Yazma ve okuma modu. Okuma ve yazma için yeni bir dosya oluşturur. Zaten mevcutsa dosyanın üzerine yazar.
### r+ : Okuma ve yazma modu. Hem okuma hem de yazma için mevcut bir dosyayı açar.
### xt : Özel metin oluşturma modu. Yazmak için yeni bir metin dosyası oluşturur, ancak zaten mevcutsa bir hata oluşturur.
### at : Metin modunu ekle. Veri eklemek için bir metin dosyası açar. (Metin dosyaları için varsayılan)
### wt : Metin modu yaz. Yazmak için yeni bir metin dosyası oluşturur. (Metin dosyaları için varsayılan)
### rt : Metin modunu oku. Okumak için mevcut bir metin dosyasını açar. (Metin dosyaları için varsayılan)
### rb : İkili modu okuyun. Okuma için mevcut bir ikili dosyayı açar.
### wb : İkili mod yazın. Yazmak için yeni bir ikili dosya oluşturur.
### ab : İkili modu ekle. Veri eklemek için ikili bir dosya açar.
### xb : Özel ikili oluşturma modu. Yazmak için yeni bir ikili dosya oluşturur, ancak zaten mevcutsa bir hata oluşturur.
