```python
import numpy as np 
```

```python
raw_data = np.random.randint(0,100,(100,5))  ## adım 1 
```

```python
subset_data = raw_data[:50 , -3:]           ## adım 2
```

```python
Filtre = subset_data[subset_data>50].mean()                 ## adım 3 
Filtre
```

```python
normalized_data = subset_data/subset_data.max(axis = 0)  ## adım 4 
```

```python
Data_t = normalized_data.T
```

```python
dot = Data_t @ normalized_data 
dot.shape
```
