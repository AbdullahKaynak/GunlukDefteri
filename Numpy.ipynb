## Lineer Regresyon

```python
## örneğimiz bir evin metrekaresı ve fiyatı olacak
X = np.array([[50], [60],[70], [80],[90], [100]])
```

```python
Y = np.array([[500], [610], [690], [820],[905],[1010]])
```

```python
## w = np.random.randn(1,1)          ## rastgele bir ağırlık atadık
w = np.array([[10]])
```

```python
## b = np.random.randn(1)            ## rastgele bir bias
b = np.array([1])
```

```python
y_pred = X @ w+b
```

```python
print(y_pred)
```

```python
error = y_pred -Y 
```

```python
mse = np.mean(error**2)
```

```python
mse
```

## burası makine öğrenmesinde aslında iş yapıyor lineer denklem

```python
# NumPy'ın lineer cebir kütüphanesini kullanarak en iyi çözümü bul
# (X matrisine 1'lerden oluşan bir sütun eklememiz gerekir bias için ama şimdilik kafan karışmasın)
from numpy.linalg import inv
```

```python
# Normal Equation (Matematiksel kesin çözüm)
# Not: Bu biraz ileri seviye bir matris işlemidir, sadece gücü gör diye veriyorum
W_optimal = inv(X.T @ X) @ X.T @ Y
print("Mükemmel Ağırlık (W):", W_optimal)
```

```python
# Şimdi bu W_optimal ile tahmin yap
y_final_pred = X @ W_optimal
final_mse = np.mean((y_final_pred - Y)**2)
print("Final Mükemmel MSE:", final_mse)
```
