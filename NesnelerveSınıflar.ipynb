## Basit bir araba sınıfını simüle eden, araba örnekleri oluşturmanıza, hızlandırmanıza ve mevcut hızlarını görüntülemenize olanak tanıyan bir python programı yazalım.

Aşağıdaki nitelikleri ve yöntemleri içeren bir Car sınıf tanımlayarak başlayalım:

120 km/s max_speed olarak ayarlanmış sınıf niteliği.

Aracın mark __init__ ası, modeli, rengi ve isteğe bağlı hız için parametreleri alan yapıcı yöntemi (varsayılan olarak 0'dır). Bu yöntem, marka, model, renk ve hız için örnek niteliklerini başlatır.

Ar accelerate(self, acceleration) acın hızlanmasını sağlayan yöntem. Hızlanma aşmazsamax_speed, aracın hız özelliğini güncelleyin. Aksi takdirde, hızı max_ speed olarak ayarlayın.

Ar get_speed(self) acın mevcut hızını döndüren yöntem.

```python
class car: 
    # Sınıf nitelikleri (tüm örnekler tarafından paylaşılan)
    max_speed = 120 # Azami hız (km/sa)

    # Yapıcı metot (örnek özelliklerini başlatır)
    def __init__(self, marka, model, renk, speed = 0):
        self.marka = marka
        self.model = model
        self.renk = renk
        self.hız = speed # Başlangıç ​​hızı 0 olarak ayarlanmıştır.

    # Arabayı hızlandırma yöntemi
    def hızlan(self,hızlanma):
        if self.hız + hızlanma <= self.max_speed:
            self.hız += hızlanma
        else:
            self.hız = car.max_speed # Maksimum hıza ulaşır

    # Aracın mevcut hızını hesaplama yöntemi
    def mevcut_hız(self):
        return self.hız
    
```

```python
car1 = car("Toyota", "Corolla", "Beyaz")
car2 = car("Honda", "Civic", "Siyah")
```

```python
car2.hızlan(50) # car2'yi 50 km/sa hızlandırır
```

```python
print(f"{car2.marka} {car2.model} şu anda {car2.mevcut_hız()} km/sa hızla gidiyor.")
```
