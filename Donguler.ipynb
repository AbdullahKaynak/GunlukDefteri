## For Döngüsü

```python
dates = [1982, 1980, 1973]
N = len(dates)

for i in range(N):
    print(dates[i])
# N kez yürütülür. ve listedekiler tek tek yazılır.
```

```python
# 8'e kadar olan ama 8 dahil olmayan sayıları yazdırır.
for i in range(8):
    print(i)
```

```python
for year in dates:
    print(year)
```

### bir liste içerisindeki öğüeleri değiştirebiliriz.

```python
squares = ['red', 'yellow', 'green', 'purple', 'blue']

for i in range(0, 5):
    print("Before square ", i, 'is',  squares[i])
    squares[i] = 'white'
    print("After square ", i, 'is',  squares[i])
```

### Bir listenin indeksine ve elemanlarına şu şekilde erişebiliriz:

```python
squares=['red', 'yellow', 'green', 'purple', 'blue']

for i, square in enumerate(squares):
    print(i, square)
```

## While Döngüsü

####  Belirtilen mantıksal koşul `False` (yanlış) değerini döndürene kadar kod bloğu çalıştırılmaya devam eder.

```python
dates = [1982, 1980, 1973, 2000]

i = 0
year = dates[0]

while(year != 1973):    
    print(year)
    i = i + 1
    year = dates[i]
    

print("Döngüden çıkmak ", i ," tekrar sürdü.")
```

## Break

#### `break` ifadesi, döngü koşuluna bakılmaksızın, içinde yer aldığı döngüyü derhal sonlandırır. 

#### Genellikle belirli bir koşul sağlandığında ve daha fazla yinelemeye gerek kalmadığında kullanılır.


#### ÖRNEK

```python
## Burada 5'i gördüğünde döngüyü kıracak ve diğer sayıları yazdırmayacak.
for num in range(1, 10):
    if num == 5:
        print("Döngüyü şurada kırma:", num)
        break
    print(num)
```

## Continue

#### continue ifadesi, döngüdeki mevcut yinelemeyi atlar ve bir sonraki yinelemeye geçer.

#### Belirli koşulları göz ardı etmek ancak döngüyü sürdürmek istediğiniz durumlarda kullanılır.

```python
## Burada 3'ü atlayacak ve diğer sayıları yazdıracak.
for num in range(1, 6):
    if num == 3:
        continue
    print(num)
```

```python
## Örnek
count = 0
while count < 10:
    count += 1
    if count == 3:
        continue
    if count == 8:
        break
    print(count)
```

#### 5-`squares` listesindeki 'orange' (portakal) dizgilerini `new_squares` listesine kopyalamak için bir `while` döngüsü yazın. Listedeki değer 'orange' değilse döngüyü durdurun ve çıkın:
