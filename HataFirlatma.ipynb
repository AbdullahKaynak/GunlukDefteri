```python
try:
    # yürütülmeye çalışılacak kod
except (ZeroDivisionError):
    # Sıfır bölme hatası oluşursa çalıştırılacak kod
except (NameError)::
    # Ad hatası oluşursa çalıştırılacak kod
except (TypeError):
    # Tür hatası oluşursa çalıştırılacak kod
except Exception as e:
    # Yukarıdaki istisnaların hiçbiri oluşmazsa çalıştırılacak kod
    print(f"Beklenmeyen bir hata oluştu: {e}")
```

## Alıştırma 1: Sıfıra Bölme Hatasını Ele Alma

### İki sayınız olduğunu ve birini diğerine böldüğünüzde ne olacağını belirlemek istediğinizi hayal edin. Bunu yapmak için, `safe_divide` adlı bir Python fonksiyonu oluşturmanız gerekir. Bu fonksiyona iki sayı, bir 'pay' ve bir 'payda' verirsiniz. 'Pay', bölmek istediğiniz sayıdır ve 'payda', bölmek istediğiniz sayıdır. Değerleri almak için Python'ın `user input` yöntemini kullanın.

```python
def safe_divide(a,b):
    try:
        result = a/b
        return result
    except ZeroDivisionError:
        print("Sıfıra bölme hatası oluştu.")
        return None 
    
a = int(input("Lütfen bölmek istediğiniz bir sayı girin:  "))
b = int(input("Lütfen bölen sayıyı girin:  "))
print(safe_divide(a,b))
    
```

## Alıştırma 2: ValueError ile Başa Çıkma

### Diyelim ki elinizde bir sayı var ve onun karekökünü hesaplamak istiyorsunuz. Bunu yapmak için bir Python fonksiyonu oluşturmanız gerekiyor. Bu fonksiyona 'sayı1' adında bir sayı veriyorsunuz.

Fonksiyon, pozitif bir tamsayı veya ondalık sayı girdisi verdiğinizde karekök değerini üretmelidir. Ancak, negatif bir değer girdiğinizde hatayı algılayacak kadar akıllı olmalıdır. Size nazikçe şu mesajı vermelidir: 'Geçersiz giriş! Lütfen pozitif bir tamsayı veya ondalık sayı girin.'

```python
import math

def sayi1(a):
    if a > 0:
        result = math.sqrt(a)
        return result
    else:
        print("Negatif bir sayı girdiniz. Lütfen pozitif bir sayı girin.")
        return None
sayi1(4)
    
```

## Alıştırma 3: Genel İstisnaların Ele Alınması

### Bir sayınız olduğunu ve karmaşık bir matematiksel işlem yapmak istediğinizi hayal edin. Hesaplama, "num" girdisinin değerini "num" ile 5 arasındaki farka bölmeyi ve sonucu "result" adlı bir değişkene kaydetmeyi gerektiriyor.

Bu karmaşık matematiksel işlemi gerçekleştirebilecek bir fonksiyon tanımlamanız gerekiyor. Fonksiyon, hesaplama sırasında oluşabilecek olası hataları ele almalıdır. Bunu yapmak için bir try-except bloğu kullanabilirsiniz. Hesaplama sırasında herhangi bir istisna oluşursa, "Exception" genel istisna sınıfını "e" olarak kullanarak hatayı yakalamalıdır. Bir istisna oluştuğunda, fonksiyon "Hesaplama sırasında bir hata oluştu." mesajını göstermelidir.

```python
def hesaplama(num):
    try:
        result = num / (num -5)
        return result
    except ZeroDivisionError:
        print("Sıfıra bölme hatası oluştu.")
        return None
    except TypeError:
        print("Geçersiz bir tür girdiniz. Lütfen sayısal bir değer girin.")
        return None
    except Exception as e:
        print(f"Beklenmeyen bir hata oluştu: {e}")
        return None
user_input = int(input("Lütfen bir sayı girin: "))
hesaplama(user_input)
```
