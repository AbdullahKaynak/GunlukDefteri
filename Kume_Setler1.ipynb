### Bir Listeyi Kümeye dönüştürelim ***************

```python
# Burada bak [] böyle liste oluşturduk.
album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
```

```python
# Burada Listeyi set komutuyla kümeye dönüştürdük. set komutu tekrarlayanları almaz
album_set = set(album_list)
album_set
```

### Kümeler üzerinde çalışma************

```python
album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

```python
album_set1, album_set2
```

## Kesişim Alma (&)

```python
intersection = album_set1 & album_set2
intersection
```

## Yalnız album1 de olanları alma(difference)

```python
album_set1.difference(album_set2)  
```

## Album1 ve Album2 kesişimleri (İntersection)

```python
album_set1.intersection(album_set2)   
```

## Album1 ve Album2 nin birleşimi (union)

```python
album_set1.union(album_set2)
```

#### Bir kümenin başka bir kümenin üst kümesi mi yoksa alt kümesi mi olduğunu şu şekilde kontrol edebilirsiniz:

```python
set(album_set1).issuperset(album_set2)   
```

```python
set({"Back in Black", "AC/DC"}).issubset(album_set1) 
```

```python
album_set1.issuperset({"Back in Black", "AC/DC"})   
```
