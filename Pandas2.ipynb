```python
import pandas as pd

### 'x' adında bir sözlük tanımlayın

x = {'Name': ['Rose','John', 'Jane', 'Mary'], 'ID': [1, 2, 3, 4], 
     'Department': ['Architect Group', 'Software Group', 'Design Team', 'Infrastructure'], 
      'Salary':[100000, 80000, 50000, 60000]}

### Sözlüğü Dataframe'e dönüştür
df1 = pd.DataFrame(x)
df1
```

### loc[satır_etiketi, sütun_etiketi]
### iloc(), indeks tabanlı bir seçme yöntemidir; yani belirli bir satır/sütunu seçmek için yönteme bir tamsayı indeksi geçirmemiz gerekir. Bu yöntem, kendisine geçirilen aralığın son öğesini içermez.
iloc[satır_indeksi, sütun_indeksi]

```python
# İlk satır ve ilk sütundaki değere erişin
print(df1.iloc[0, 0])

# Birinci satırdaki ve üçüncü sütundaki değere erişin
print(df1.iloc[0, 2])

# Sütuna adını kullanarak erişin
print(df1.loc[0, "Salary"])
```

#### 'df2' adında yeni bir veri çerçevesi oluşturalım ve 'df'yi ona atayalım. Şimdi, set_index() yöntemini kullanarak "Student" sütununu indeks sütunu olarak ayarlayalım.

```python
df2 = df1
df2 = df2.set_index("Name")
```

```python
#Yeni veri çerçevesinin ilk 5 satırını görüntülemek için
df2.head()

# Şimdi, sütuna adını kullanarak erişelim.
df2.loc["Jane", "Salary"]
```

```python
# Yeni oluşturulan df2 veri çerçevesinde Jane'in bölümünü almak için loc() fonksiyonunu kullanın.
df2.loc["Jane", "Department"]
```

```python
## Yeni oluşturulan df2 veri çerçevesinde Mary'nin maaşını almak için iloc() fonksiyonunu kullanın.
df2.iloc[3, 2]
# df2.loc["Mary", "Salary"]
```

## Alıştırma 3: Dilimleme
Bir dizi satırı dilimlemek için şu sözdizimini kullanırsınız: data[başlangıç:bitiş]

```python
## Eski veri çerçevesi df'yi kullanarak dilimleme işlemini yapalım
df1.iloc[0:2, 0:3 ]
```

```python
#Eski veri çerçevesi df üzerinde loc() fonksiyonunu kullanarak dilimleme işlemini yapalım; 
# burada index sütununda 0, 1, 2 etiketleri bulunmaktadır.
df1.loc[0:2, "ID": "Department"]
```

```python
## Yeni df2 veri çerçevesinde, indeks sütunu "Name" olan ve etiketleri "Rose", "John" ve "Jane" olan bir alanda,
#  loc() fonksiyonunu kullanarak dilimleme işlemini yapalım.
df2.loc["Rose":"Jane", "ID": "Department"]
```
