# Writing Files
Metin dosyasını kaydetmek için `write()` metodunu kullanarak bir dosya nesnesi açabiliriz. Bir dosyaya yazmak için `mode` argümanı `w` olarak ayarlanmalıdır. Şimdi, içinde “This is line A” satırı bulunan `Example2.txt` adlı bir dosya oluşturalım.

```python
exmp2 = 'example2.txt'    ## hangi dosyaya yazmak istiyorsak 
with open(exmp2, "w") as writefile:         ## dosyaya yazalım 
    writefile.write("This is line A")
```

```python
with open(exmp2, "w") as writefile:
    writefile.write("This is line A\n")
    writefile.write("This is line B\n")
```

```python
## Dosya içeriğini okumak
with open(exmp2, "r") as testwirtefile:
    print(testwirtefile.read())
```

```python
## bir listeyi .txt dosyasına yazmak
Lines = ["This is line A\n", "This is line B\n", "This is line C\n"]
Lines
with open("examplle.txt", "w") as writefile:        ## examplle adında bir text oluşturduk ve içine yazdık
    for line in Lines:
        print(line)
        writefile.write(line)
```

```python
## Dosya okuma yolları

with open("examplle.txt", "r") as okuma:
    print(okuma.read())

with open("examplle.txt", "r") as testwirtefile:
    print(testwirtefile.read())
```

# Dosya Ekleme (Append)
Mevcut verilerin hiçbirini kaybetmeden dosyalara şu şekilde yazabiliriz: Bunun için mod argümanını "append" olarak ayarlamanız yeterlidir: a. Yeni bir satır eklemek için aşağıdaki adımları izleyebilirsiniz:

```python
with open("examplle.txt", "a") as testwirtefile:
    testwirtefile.write("This is line C\n")
    testwirtefile.write("This is line D\n")
    testwirtefile.write("This is line E\n")
```

```python
with open("examplle.txt","r") as okuma:
    print(okuma.read())
```

## Ek modlar¶

`r+`: Okuma ve yazma. Dosyayı kısaltamaz.

`w+`: Yazma ve okuma. Dosyayı kısaltır.

`a+`: Ekleme ve okuma. Eğer dosya yoksa yeni bir dosya oluşturur. 

`.tell()` - Şu an imleç kaçıncı baytta/karakterde?" sorusunu sorar.

`.seek(0,0)` - İmleci dosyanın en başına (0. noktaya) götür" komutudur. (x ve y notkaları)

```python
# append+
with open("examplle.txt", "a+") as testwirtefile:
    print("Başlangıç ​​Konumu: {}".format(testwirtefile.tell()))

    data = testwirtefile.read()
    if (not data):
        print("Hiçbir şey okuma")
    else: 
        print(testwirtefile.read())

    
    testwirtefile.seek(0) ## imleci başa aldık
    print("\nYeni konum: {}".format(testwirtefile.tell()))
    data = testwirtefile.read()

    if(not data):
        print("Hiçbir şey Okuma")
    else:
        print(data)

    print("sonraki okuma konumu: {}". format(testwirtefile.tell()))
```

`w+` ve `r+` arasındaki farka değinelim. Her iki mod da okuma ve yazma yöntemlerine erişime izin verir; ancak `w+` modunda bir dosya açmak, dosyayı üzerine yazar ve önceden var olan tüm verileri siler.

```python
with open("examplle.txt", "r+") as testwirtefile:
    testwirtefile.seek(0,0)   #dosyanın başına yaz
    testwirtefile.write("Line 1" + "\n")
    testwirtefile.write("Line 2" + "\n")
    testwirtefile.write("Line 3" + "\n")
    testwirtefile.write("Line 4" + "\n")
    testwirtefile.write("Line 5" + "\n")
    testwirtefile.write("Line 6" + "\n")
    testwirtefile.write("Finished")
    testwirtefile.seek(0,0)
    print(testwirtefile.read())
```

Mevcut verilerle bir dosya üzerinde çalışmak için `r+` ve `a+` kullanın. `r+` kullanırken, verilerinizin sonuna `.truncate()` yöntemini eklemek faydalı olabilir. Bu, dosyayı verilerinize indirger ve sonrasındaki her şeyi siler.

```python
with open("examplle.txt", "r+") as testwirtefile:
    testwirtefile.seek(0,0) 
    testwirtefile.write("Line 1" + "\n")
    testwirtefile.write("Line 2" + "\n")
    testwirtefile.write("Line 3" + "\n")
    testwirtefile.write("Line 4" + "\n")
    testwirtefile.write("finished\n")
    testwirtefile.truncate()
    testwirtefile.seek(0,0)
    print(testwirtefile.read())
```

# Dosya Kopyala
`Examplle.txt` dosyasını `Example3.txt` dosyasına kopyalayalım:

```python
with open("examplle.txt", "r") as testwirtefile:
    with open("example3.txt", "w") as writefile:
        for line in testwirtefile:
            writefile.write(line)
```

```python
with open("example3.txt", "r") as writefile:
    print(writefile.read())
```
