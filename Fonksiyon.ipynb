## Toplama Fonksiyonu oluşturma

```python
def calculate_total(a,b):
    total = a+ b
    return total

result = calculate_total(5, 7 )
print(result)
```

## Len, Sum, Max, Min Örnekleri

```python
# Len kullanımı bir dizinin koleksiyon uzunluğunu hesaplar
string_length = len("Hello, World!")
print(string_length)

# Sum toplama işlemleri için kullanılır.
total = sum([1, 2, 3, 4, 5])
print(total)

# Max ve min fonksiyonları bir dizideki en büyük ve en küçük değerleri bulmak için kullanılır.
numbers = [3, 7, 1, 9, 4]
max_value = max(numbers)
min_value = min(numbers)
print(max_value)
print(min_value)
```

### Fonksiyonları kullanarak veri yapısını değiştirme

```python

def add_element(data_structure,element):
    data_structure.append(element)

def remove_element(data_structure,element):  ##  Data_structure bir öğe eklemek istediğimiz listeyi temsil eder. element ise eklemek istediğimiz öğeyi temsil eder.
    if element in data_structure:
        data_structure.remove(element)
    else:
        print(f"{element} listede bulunamadı.")
my_list = []
add_element(my_list, 42)
add_element(my_list, 17)            ### Burada listeye elemen ekliyoruz.
add_element(my_list, 99)
print("Current list:", my_list)

remove_element(my_list, 17)          ## Burada listeden eleman çıkarıyoruz.
print("Updated list:", my_list)

print("Final List:" , my_list)
```
