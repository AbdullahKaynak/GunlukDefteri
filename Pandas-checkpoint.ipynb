```python
df1 = pd.read_csv('onlinefoods.csv')
```

```python
df1.head()
```

```python
df1.groupby('Gender')['Age'].agg(['count', 'mean'])
```

```python
df1.rename(columns={'Gender': 'Cinsiyet'}, inplace=True)
```

```python
df1.head()
```

```python
for Gender,grup in df1.groupby('Cinsiyet'):
    print(f"---{Gender} Analizi ---")
    print(f"Kişi Sayısı: {len(grup)}")
    print(f"Yaş Ortalaması {grup['Age'].mean():.2f}")
    print(f"En sık Meslek {grup['Occupation'].mode()[0]}")
    print("-"*20)
```

```python
df1.drop(labels=['Unnamed: 12'], axis=1, inplace=True)
```

```python
df1.head()
```

```python
obj_cols = df1.select_dtypes(['object']).columns
for col in obj_cols:
    df1[col] = df1[col].str.strip()
```

```python
df1.head()
```

```python
income_map = {
    'No Income':0,
    'Below Rs.10000': 5000,
    '10000 - 25000': 17500,
    '25000 - 50000': 37500,
    'More than 50000': 60000
}
df1['Monthly_Income_Numeric'] = df1['Monthly Income'].map(income_map)
```

```python
df1.head()
```

```python
df1.groupby('Occupation')['Monthly_Income_Numeric'].mean().sort_values(ascending=False)
```
