```python
import numpy as np
```

```python
data = np.random.randint(0,1000, (1000,4))
data.shape
```

```python
features = data[::, :2]
features.shape
```

```python
vip_customers = features[features[:: , 0]>900]
```

```python
ortalama = vip_customers.mean(axis = 0)
centered_vip = vip_customers-ortalama 
```

```python
dot = centered_vip.T
```

```python
yeni = dot@centered_vip
yeni.shape
```
