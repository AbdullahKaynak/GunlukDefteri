# Dataframe
DataFrame, potansiyel olarak farklı veri türlerinin sütunlarına sahip iki boyutlu etiketli bir veri yapısıdır.

DataFrames sözlüklerden oluşturulabilir, anahtarlar sütun etiketleri ve değerler satırları temsil eden listeler olarak oluşturulabilir.

```python
data = {"Name": ["Alice", "Bob", "Charlie", "David"],
        "Age": [25,30,35,28],
        "City": ["New York", "San Francisco", "Los Angeles", "Chicago"]
        }

df = pd.DataFrame(data)
print(df)
```

## Sütun seçimi

```python
print(df["Name"])
```

## Satırlara erişim

```python
print(df.iloc[2])   # 2.DEĞERİ KOMPLE DÖNDÜRÜR
```

```python
print(df.loc[2])
```

## Dilimleme

```python
print(df[["Name", "Age"]])  ## Name ve Age kısımlarını komple getirir (Columns)      
```

```python
print(df[1:3])  #rows
```

```python
unique_dates = df["Age"].unique()       ## yaş sütünlarını bir dizeye saklar benzersiz kodları  
unique_dates
```

## Koşullu Filtreleme

```python
hight_above_102 = df[df["Age"] > 25]
hight_above_102
```

## Veri Çerçevelerini kaydetme

```python
df.to_csv('trading_data.csv', index=False)
```
