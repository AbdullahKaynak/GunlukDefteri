## Parametre a'ya ekleme yapan ve çıktıyı b olarak yazdırıp döndüren bir fonksiyon örneği:

```python
def add(a):
    """
    a değerine 1 ekleme fonksiyonu
    """
    b = a + 1
    return(b)
# help(add) -> burada fonksiyonun açıklamasını görebilirsin. 3 tırnak içerisine yaptığın açıklama gelir önümüze.
add(6)
```

## Sayı çarpım fonksiyonu

```python
def carpim(a,b):
    """
    a ve b değerlerinin çarpımını döndüren fonksiyon
    """
    c = a*b
    return(c)
    print("This is not printed")  # return'den sonra yazılan kodlar çalışmaz. return fonksiyonu fonksiyonu bitirir.

result = carpim(3,4)
result = carpim(30.5, 2.5)
result = carpim(3, 4) + carpim(5, 6)
result = carpim(3, "Abdullah ")  # string ile çarpma işlemi yapılabilir. stringi n kadar çarpacağını belirtir.
print(result)
```

### Fonksiyonlarda if/else Deyimleri ve Döngülerin Kullanımı

## Sözlük Kullanarak Bir Dizideki Kelime Sayısını Hesaplayan Python Programı

```python
def freq(string):
    
    ## Step 1 : Boş bir liste oluştur.
    words = []

    ## Step 2: Dizeyi kelime listesine ayırın.
    words = string.split()

    ## Step 3: Boş bir sözlük oluşturun.
    Dict = {}

    ## Step 4: Sözlükteki her kelimeyi sayın.(döngü oluştur)
    for key in words:
        Dict[key] = words.count(key)

    ## Step 5: Sözlüğü döndürün.
    print("The Frequency of words is: ", Dict)

## Adım 6: Fonksiyonu çağırın ve içine dizeyi iletin.
freq("Mary'nin küçük bir kuzusu vardı. Küçük kuzu, küçük kuzu. Mary'nin küçük bir kuzusu vardı. Yünü kar gibi beyazdı. \
    Ve Mary nereye giderse gitsin, kuzu da mutlaka onunla giderdi. Mary nereye giderse gitsin, kuzu da mutlaka onunla giderdi.")


```

## Özel fonksiyonlarınızda varsayılan argüman değerlerini ayarlama
Fonksiyonunuzdaki argümanlar için varsayılan bir değer ayarlayabilirsiniz. Örneğin, isGoodRating() fonksiyonunda, iyi bir puan olarak kabul ettiğimiz bir eşik oluşturmak isteseydik ne olurdu? Belki de varsayılan olarak 4'lük bir puanımız olmalıdır:

```python
def isGoodRating(rating = 4):
    if(rating < 7):
         print("Bu albüm berbat, puanı da düşük.", rating)          ## böyle print yazdırırsan none döner. return ile döndürürsen None dönmez.
    else:
        return f"Bu albüm iyi, puanı da yüksek.{rating}" 

print(isGoodRating())
print(isGoodRating(10))
```

## Bir fonksiyonun argüman sayısı bilinmediğinde, bunların tümü aşağıdaki gibi bir demet (tuple) içine yerleştirilebilir:

```python
def printAll(*args): # Tüm argümanlar, bir demet gibi ele alınabilen 'args' içine 'paketlenmiştir'.
    print("No of arguments:", len(args)) 
    for argument in args:
        print(argument)
#printAll fonksiyonuna 3 argüman atanır.
printAll('Horsefeather','Adonis','Bone')
#printAll fonksiyonuna 4 argüman atanır.
printAll('Sidecar','Long Island','Mudslide','Carriage')
```

```python
# Similarly, The arguments can also be packed into a dictionary as shown:

def printDictionary(**args):
    for key in args:
        print(key + " : " + args[key])

printDictionary(Country='Canada',Province='Ontario',City='Toronto')
```
