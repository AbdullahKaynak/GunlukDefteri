Bir sonraki adım, nesneyi başlatmak için kullanılan __init__ adlı özel bir kurucu metottur. Girişler veri öznitelikleridir. `self` terimi, kümedeki tüm öznitelikleri içerir. Ayrıca, `r` parametresiyle `add_radius()` metodumuz da vardır; bu metot, `r` değerini `radius` özniteliğine ekler. Yarıçapa erişmek için `self.radius` sözdizimini kullanırız.

## Şekil 5: Daire nesnesinin etiketlenmiş sözdizimi.

Aşağıda gerçek nesne gösterilmektedir. Dairenin görüntüsünü göstermek için drawCircle yöntemini ekledik. Varsayılan yarıçapı 3 ve varsayılan rengi mavi olarak ayarladık:

```python
# Bir Circle sınıfı oluşturalım.

class Circle(object):

    ## Yapıcı Method
    def __init__(self, radius=3, color ="blue"):
        self.radius = radius
        self.color = color
    
    ## Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return self.radius
    
    ## Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0,0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()

```

# Circle sınıfının bir örneğini oluşturma
Aşağıdakileri yapmak için Circle türünde RedCircle adlı bir nesne oluşturalım:

```python
# Kırmızı Daire nesnesi oluşturun

RedCircle = Circle(10, "red")
```

#### Nesnenin metotlarının listesini almak için `dir` komutunu kullanabiliriz. Bunların çoğu varsayılan Python metotlarıdır.

```python
dir(RedCircle)
```

#### Nesnenin veri özelliklerine bakabiliriz:

```python
print(RedCircle.radius)
print(RedCircle.color)
```

#### Nesnenin veri özelliklerini değiştirebiliriz:

```python
# Nesne özniteliği yarıçapını ayarlayın

RedCircle.radius = 1
RedCircle.radius
```

#### drawCircle() yöntemini kullanarak nesneyi çizebiliriz:

```python
RedCircle.drawCircle()
```

#### `add_radius()` metodunu kullanarak dairenin yarıçapını artırabiliriz. Önce yarıçapı 2, sonra da 5 artıralım:

```python
# Nesnenin yarıçap özelliğini değiştirmek için yöntemi kullanın

print("Nesneğin yarıçapı: ", RedCircle.radius)

RedCircle.add_radius(2)
print("İlk yöntemi uyguladıktan sonra nesnenin yarıçapı: ", RedCircle.radius)

RedCircle.add_radius(5)
print("İkinci yöntemi uyguladıktan sonra nesnenin yarıçapı: ", RedCircle.radius)
```

### Mavi bir daire oluşturalım. Varsayılan renk mavi olduğundan, yapmamız gereken tek şey yarıçapı belirtmek:

# Dikdörtgen Sınıfı
Yükseklik, genişlik ve renk özelliklerine sahip bir dikdörtgen sınıfı oluşturalım. Sadece dikdörtgen nesnesini çizecek metodu ekleyeceğiz:

```python
class Rectangle(object):

    # Yapıcı metod
    def __init__(self, width=2, height =3, color= "r"):
        self.height = height
        self.width = width
        self.color = color
    
    # metod
    def drawRenctangle(self):
        plt.gca().add_patch(plt.Rectangle((0,0), self.width, self.height, fc = self.color))
        plt.axis("scaled")
        plt.show()
```

### Şimdi Rectangle türünde SkinnyBlueRectangle adlı bir nesne oluşturalım. Genişliği 2, yüksekliği 3 ve rengi mavi olacak:

```python
SkinnyBlueRectangle = Rectangle(2,3 , "blue")
```

```python
# Daha önce olduğu gibi, nokta gösterimini kullanarak sınıf örneğinin özelliklerine erişebiliriz:

print(SkinnyBlueRectangle.height)
print(SkinnyBlueRectangle.width)
print(SkinnyBlueRectangle.color)
SkinnyBlueRectangle.drawRenctangle()
```

### Şimdi Rectangle türünde FatYellowRectangle adlı bir nesne oluşturalım:

```python
FatYellowRectangle = Rectangle(20,5, "yellow")
print(FatYellowRectangle.width)
print(FatYellowRectangle.height)
print(FatYellowRectangle.color)
FatYellowRectangle.drawRenctangle()
```

# Senaryo: Otomobil bayisinin stok yönetim sistemi 
Bir otomobil bayisinin envanter yönetim sistemini simüle etmek için bir Python programı üzerinde çalışıyorsunuz. Sistem, otomobilleri ve özelliklerini doğru bir şekilde modellemeyi amaçlıyor.

### Görev 1. Araçları bir sınıf kullanarak temsil eden bir Python programı oluşturmanız gerekiyor. Her aracın maksimum hız ve yakıt tüketimi için nitelikleri olmalıdır.
### Görev-2. Tüm araçlar için varsayılan renk olan "beyaz"ı kullanarak sınıfı güncelleyin.
### Görev-3. Ayrıca, bir araca oturma kapasitesi atamak için Vehicle sınıfında yöntemler oluşturmanız gerekmektedir.
### Görev 4. Sınıfın bir nesnesinin tüm özelliklerini görüntüleyen bir yöntem oluşturun.
### Görev-5. Ek olarak, azami hızı 200 km/sa ve yakıt tüketimi 20 km/litre olan, beş koltuk kapasiteli iki adet Araç sınıfı nesnesi ve azami hızı 180 km/sa ve yakıt tüketimi 25 km/litre olan, dört koltuk kapasiteli bir adet Otomobil nesnesi oluşturmanız gerekmektedir.

```python
class car:
    color = "White"              ##  Görev 2

    # Yapıcı Görev 1
    def __init__(self, max_hiz, yakit):
        self.max_hiz = max_hiz
        self.yakit = yakit
        self.oturma_kapasitesi = None
    
    def oturma_kapasitesi_atama(self, oturma_kapasitesi):     ## Görev 3 
        self.oturma_kapasitesi = oturma_kapasitesi

    def goruntulenme_ozellikleri(self):                         ### Görev 4 
        print("Aracın Özellikleri: ")
        print("Renk: ", self.color)
        print("Maksimum hız: ", self.max_hiz)
        print("Kilometre: ", self.yakit)
        print("Oturma Kapasitesi: ", self.oturma_kapasitesi)
```

```python
# Görev 5

Arac1 = car(200,20)
Arac1.oturma_kapasitesi_atama(5)
Arac1.goruntulenme_ozellikleri()
print("**************************")
Arac2 = car(180,25)
Arac2.oturma_kapasitesi_atama(4)
Arac2.goruntulenme_ozellikleri()

```
